```python
import pandas as pd

df = pd.read_csv("dataset_trot_attele.csv", parse_dates=['date_course'])

display(df.head())
print(f"Dimensions du dataset : {df.shape}")
```

```python
categorical_cols = ['hippodrome', 'shoeing_status', 'driver', 'entraineur', 'sex']

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')


df.info()
```
