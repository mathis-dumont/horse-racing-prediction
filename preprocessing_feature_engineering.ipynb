## Preprocessing

### Import des données


```python
import pandas as pd
import numpy as np




pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)



data = pd.read_csv("dataset/export-1-pt-utf.csv")

data.head()
```

### Nettoyage

```python
data.info()






```

#### on renomme les colonnes où ya des espaces

```python
data = data.rename(columns={
    "cr-etat du terrain": "cr_etat_terrain",
    "cr-corde": "cr_corde",
})
```

```python
data["cr-Date"] = pd.to_datetime(data["cr-Date"], errors="coerce")
# on convertit en datetime
```

```python
data.isna().mean().sort_values(ascending=False)
# on regarde le pourcentage de valeurs manquantes par colonne
```

```python
X = data.drop(columns=["place"])
```

```python
# 1. On part de data déjà nettoyé des colonnes évidentes
cols_to_drop = [
    "placeoptin", "cr-Date", "cr-reunion",
    "cr-num", "Numcheval", "NumCourseJockey",
    "SumPoids", "SumIdJockey", "SumidHippodrome",
    "SumIdDeferre", "SumDistance", "SumAllocation",
    "SommeCourse", "rapport"
]

data = data.drop(columns=cols_to_drop, errors="ignore")

```

#### on encode les variables catégorielles

```python
cat_cols = ["cr-Hippodrome", "cr-Evt", "cr_corde", "cr_etat_terrain", "Sexe"]

data = pd.get_dummies(data, columns=cat_cols, drop_first=True)
```

```python
# On extrait la partie numérique de la chaîne ('2.725' dans '2.725 mètres')
data["cr-distance"] = (
    data["cr-distance"]
        .str.extract(r"([\d\.]+)")[0]   # on garde juste les chiffres + point
        .astype(float)
)
```

#### CoteProbable → on ajoute sa version log-transfo → les cotes sont très "skewed" : la transformation les rend plus faciles à utiliser.

```python
data["log_cote"] = np.log1p(data["CoteProbable"])
```

```python
list(data.columns)
```

on construit x et y

```python
X = data.drop(columns=["place"])
y = data["place"]


#  Suppression des colonnes constantes (sur TOUT le jeu)

const_cols = X.nunique()[X.nunique() == 1].index.tolist()
X = X.drop(columns=const_cols)

print("Colonnes constantes supprimées :", len(const_cols))
```

## Importance des features

on va regarder ça avec random forest

```python
data.shape
```

```python
data["place"].unique()
```

```python
y.value_counts(normalize=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=8
)      # stratify car classes déséquilibrées
```

random forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    n_jobs=-1,
    random_state=8,
    # class_weight="balanced"  # optionnel ici, 55/45 c'est pas un gros déséquilibre
)

rf.fit(X_train, y_train)

```

on regardes les perfs

```python
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score

y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]

print("Accuracy :", accuracy_score(y_test, y_pred))
print("AUC ROC  :", roc_auc_score(y_test, y_proba))
print("\nClassification report :\n", classification_report(y_test, y_pred))
```
